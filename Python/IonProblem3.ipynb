```python
import plotly.graph_objects as go
import math
import numpy
```

```python
def createParticle(init, x):
    return {
        "x_initial" : init,
        "x" : x,
        "v" : 0.0,
        "a" : 0.0,
        "history" : []
    }
```

```python
particles = [
    createParticle(4 * pow(10,-6), pow(10,-7)),
    createParticle(-4 * pow(10,-6), -pow(10,-7)),
]
```

```python
freq = 1.0 * pow(10,7)
w = freq * 2.0 * math.pi
m = 9*(1.6e-27)
omega = math.sqrt(3.0)*w

time_end = (2.0*math.pi/w) * 10.0

# I blanked out on what to put here so I just guessed. 
k = (8.988 * pow(10, 9)) * 1
q = (1.602 * pow(10,-19))

dt = (1/omega) * pow(10, -2)
# print(omega)


particles[0]["x_initial"] = pow((k*q*q) / (4 * m * pow(w,2)), 1/3)
particles[1]["x_initial"] = -1 * particles[0]["x_initial"]

time_data = []

def getSeries(wx):
    global particles
    global m
    global q
    global dt
    global w
    global omega
    global q
    global k
    global dt
    s_max_0 = float('-inf')
    s_min_0 = float('inf')
    s_max_1 = float('-inf')
    s_min_1 = float('inf')
    particles = [
        createParticle(4 * pow(10,-6), 0),
        createParticle(-4 * pow(10,-6), 0),
    ]
    particles[0]["x_initial"] = pow((k*q*q) / (4 * m * pow(w,2)), 1/3)
    particles[1]["x_initial"] = -1 * particles[0]["x_initial"]
    for t in numpy.arange(0, time_end, dt):
        particles[0]["a"] = ((q * pow(10,-1) *math.cos(wx* t)) / m)
        particles[1]["a"] = ((q * pow(10,-1) *math.cos(wx * t)) / m)
        particles[0]["a"] = ((((-1 * m) * pow(w, 2) * (particles[0]["x"] + particles[0]["x_initial"])) + ((k*q*q) / pow((particles[1]["x"] + particles[1]["x_initial"]) - (particles[0]["x"] + particles[0]["x_initial"]), 2))) /m) + ((q* pow(10,-2) *math.cos((wx)* t)) / m)
        particles[1]["a"] = ((((-1 * m) * pow(w, 2) * (particles[1]["x"] + particles[1]["x_initial"])) - ((k*q*q) / pow((particles[1]["x"] + particles[1]["x_initial"]) - (particles[0]["x"] + particles[0]["x_initial"]), 2))) /m) - ((q * pow(10,-2) *math.cos((wx) * t)) / m)
        # time_data.append(t)
        particles[0]["v"] += particles[0]["a"] * dt
        particles[1]["v"] += particles[1]["a"] * dt
        particles[0]["x"] += particles[0]["v"] * dt
        particles[1]["x"] += particles[1]["v"] * dt
        particles[0]["history"].append(particles[0]["x"])
        particles[1]["history"].append(particles[1]["x"])
        s_min_0 = min(s_min_0, particles[0]["x"])
        s_min_1 = min(s_min_1, particles[1]["x"])
        s_max_0 = max(s_max_0, particles[0]["x"])
        s_max_1 = max(s_max_1, particles[1]["x"])
    return [{
        "min":s_min_0,
        "max":s_max_0
    }, {
        "min":s_min_1,
        "max":s_max_1
    }]


wr = 2.0*math.pi*pow(10, 7)
dw = 2.0*math.pi*pow(10, 5)

p0 = []
p1 = []

wcent = w*pow(3.0,0.5)

for wcur in numpy.arange(wcent-wr, wcent+wr, dw):
    time_data.append(wcur/wcent)
    print(wcur/wcent)
    series = getSeries(wcur)
    p0.append(series[0]["max"] - series[0]["min"])
    p1.append(series[1]["max"] - series[1]["min"])




fig = go.Figure(data=[go.Line(y=p0, x=time_data), go.Line(y=p1, x=time_data)])
fig.show()
```
